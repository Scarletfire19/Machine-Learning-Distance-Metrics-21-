```python
from scipy.spatial import distance

import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

Vector Distance Metrics

```python
df=pd.read_csv('/content/G25IIT1.csv')
df.head()
```

```python
dfnext=df.groupby(['DNA sample ethnicity']).mean().reset_index()
```

```python
dfnext25 = dfnext.iloc[:,1:]
```

```python
#euclidean
deuclidean=distance.cdist(dfnext25,dfnext25, metric='euclidean')
dmateuclidean=pd.DataFrame(deuclidean)
print(dmateuclidean)
figeuclidean=sns.clustermap(dmateuclidean,figsize=(20, 30))
```

```python
#braycurtis
dbraycurtis=distance.cdist(dfnext25,dfnext25, metric='braycurtis')
dmatbraycurtis=pd.DataFrame(dbraycurtis)
print(dmatbraycurtis)
figbraycurtis=sns.clustermap(dmatbraycurtis,figsize=(20, 30))
```

Set Metrics

```python
input=dfnext25
```

```python
#jaccard
pjaccard = np.zeros((len(input),len(input)))
for i in range(len(input)):
  for j in range(len(input)):
    pjaccard[i][j]=distance.jaccard(input.loc[[j]],input.loc[[i]])          
  djaccard=pd.DataFrame(pjaccard)
print(djaccard)
figdjaccard=sns.clustermap(djaccard,figsize=(20, 30))
```

```python
#dice
pdice = np.zeros((len(input),len(input)))
for i in range(len(input)):
  for j in range(len(input)):
    pdice[i][j]=distance.dice(input.loc[[j]],input.loc[[i]])          
  distmatdice=pd.DataFrame(pdice)
print(distmatdice)
fig11=sns.clustermap(distmatdice,figsize=(20, 30))
```
